```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")
```

```python
Male_survivors=df[(df["Survived"]==1)&(df["Sex"]=="male")]
Female_survivors=df[(df["Survived"]==1)&(df["Sex"]=="female")]
```

```python
Male_survivors.plot(x="Age",y="Fare",kind="scatter",color="g",label="male",title="Male Survivors Age and Fare",figsize=(10,8))
plt.legend()
plt.show()
```

```python
Female_survivors.plot(x="Age",y="Fare",kind="scatter",color="r",label="female",title="Female Survivors Age and Fare",figsize=(10,10))
plt.legend()
plt.show()
```
