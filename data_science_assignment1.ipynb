```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")
```

```python
Male_survivors=df[(df["Survived"]==1)&(df["Sex"]=="male")]
Female_survivors=df[(df["Survived"]==1)&(df["Sex"]=="female")]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(Male_survivors["Age"],Male_survivors["Fare"],c="g",label="Male Survivors",marker="o")
plt.scatter(Female_survivors["Age"],Female_survivors["Fare"],c="r",label="Female Survivors",marker="o")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Survivors and Fare")
plt.legend(title="Survivors")
plt.show()
```
